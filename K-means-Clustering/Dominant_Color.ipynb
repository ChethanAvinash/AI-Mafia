```python
import matplotlib.pyplot as plt
import numpy as n
import cv2
from sklearn.cluster import KMeans
```

```python
img=cv2.imread("city.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
original_img=img
```

```python
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
pixels=img.reshape((1282*2278,3))
```

```python
pixels.shape
```

```python
k=4
km=KMeans(n_clusters=4)
km.fit(pixels)
```

```python
h=km.cluster_centers_
h=n.array(h,dtype="uint8")
```

```python
colors=[]
j=1
for i in h:
    plt.subplot(1,4,j)
    plt.axis("off")
    j+=1
    colors.append(i)
    f=n.zeros((100,100,3),dtype="uint8")
    f[:,:,:]=i
    plt.imshow(f)
plt.show()
```

```python
p=km.labels_
```

```python
len(p)
```

```python
new_img=n.zeros((2920396,3),dtype="uint8")
new_img
```

```python
for i in range(new_img.shape[0]):
    new_img[i]=colors[km.labels_[i]]
new_img=new_img.reshape((original_img.shape))
plt.imshow(new_img)
plt.show()
```
